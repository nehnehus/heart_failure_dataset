```python
#Histogram chart of Cholesterol
plt.figure(figsize=(10,6))
sns.histplot(heart_failure['Cholesterol'],bins=30, kde=True, color='green')
plt.title('Distribution for Cholesterol')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
```

```python
#checking the value counts of heart disease
target_counts = heart_failure['HeartDisease'].value_counts()
target_counts
```

```python
#Comparing Heart Disease By the Gender(male, female)
comparison_data = heart_failure.groupby(['HeartDisease','Sex']).size().unstack()
comparison_data.plot(kind='bar', figsize=(8,6), color=['pink','lightblue'])
plt.xticks(rotation=0)
plt.ylabel("Patient Count", fontsize=12)
plt.title("Heart Disease Comparison by Gender")
plt.show()
```

```python
#Comparison Between The Age and Maximum Heart Rate
heart_failure.plot(kind='scatter', x='Age', y='MaxHR', alpha=0.7, c='HeartDisease', cmap='coolwarm', figsize=(10,6))
plt.title('Relationship Between Age and Max Heart Rate')
plt.xlabel('Age(Years)')
plt.ylabel('Max Heart Rate')
plt.show()
```
